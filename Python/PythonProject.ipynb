<h2>Machine learning: classification</h2>

could our variables of interest predict student sucess rate? (graduate or dropout)

```python
# housekeeping 
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error

from sklearn.ensemble import RandomForestClassifier

# getting seaborn online to give our graphics a kick.
import seaborn as sns
sns.set()
```

```python
# The outcome variable is a str variable and it should be encoded
from sklearn.preprocessing import LabelEncoder # if you have categorical variables that needs to be encoded use this
le = LabelEncoder()
df_gradORdropout.loc[:, 'target_enc'] = le.fit_transform(df_gradORdropout['Target'])
```

```python
# Python libraries like Scikit expect the data to the form of 2 different components: 
    # (1) learn patterns from features (predictors) 
    # (2) use features to predict labels (outcome) 
# this is why we need to split the data into two separate ones 

# print(df_gradORdropout.columns)

# Separating independent and dependent variables (see above)
# Assuming df_gradORdropout is your DataFrame
X = df_gradORdropout[[
    'Marital status', 
    'Course', 
    'Daytime/evening attendance', 
    'Previous qualification', 
    "Mother's qualification", 
    "Father's qualification", 
    "Mother's occupation", 
    "Father's occupation", 
    'Displaced', 
    'Educational special needs', 
    'Debtor', 
    'Gender', 
    'Scholarship holder', 
    'Age at enrollment'
]]
 # you can either drop all irrelevant columns or explicitly include all relevant ones (*just make sure the outcome variable of interest is not included)
Y = df_gradORdropout['target_enc'] 


# checking 
print(X.columns)
print(Y)
```

<h3>Splitting training and test data</h3>

```python
# Splitting - 20% used for testing and 80% used to train the model
X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.2, # In this case, test_size=0.2 means 20% of the data will be used for testing, and the remaining 80% will be used for training.
                                                    random_state=42, # this ensures that the split is reproducible—i.e., every time you run the code, you get the same split. You can use any integer, but using 42 is common in practice for consistency.
                                                    stratify=Y) # i have an imbalance outcome distribution so this ensures that the class distribution in Y (your target) is maintained in both the training and test sets.

# Creating the classifier object.
RFR = RandomForestClassifier(n_estimators=500, # this specifies the number of decision trees the Random Forest will build
                             max_depth=500, # this specifies the max depth per tree The depth of a tree refers to how many splits or levels it can make before reaching a leaf node (where classification happens). Deeper trees can capture more details from the training data, but they can also lead to overfitting.
                             random_state=42)  # this ensures that the split is reproducible—i.e., every time you run the code, you get the same split. You can use any integer, but using 42 is common in practice for consistency.

# Fitting the train dataset.
RFR.fit(X_train,Y_train)

# Creating the predictions.
Y_pred = RFR.predict(X_test)

# Calculating error
print(mean_absolute_error(Y_test,Y_pred))
```

<h1>Secondary analysis</h1>

My secondary question: could macro-economic variables predict courses enrolled? 

The current course category is very specific and I'm not interested in whether our variables can predict specific courses but rather the broader category so let's regroup them into 6 broad categories and collapse the evening courses with the normal ones: 

1. **Engineering and Technology**
    - 1—Biofuel Production Technologies
    - 7—Informatics Engineering
  
2. **Arts, Design, and Multimedia**
    - 2—Animation and Multimedia Design
    - 5—Communication Design

3. **Agriculture and Veterinary Sciences**
    - 4—Agronomy
    - 6—Veterinary Nursing
    - 8—Equiniculture
    
4. **Social Sciences and Services**
    - 3—Social Service (evening attendance)
    - 10—Social Service
    - 16—Basic Education

5. **Business and Management**
    - 9—Management
    - 11—Tourism
    - 14—Advertising and Marketing Management
    - 17—Management (evening attendance)
      
6. **Health Sciences**
    - 12—Nursing
    - 13—Oral Hygiene
      
7. **Media and Communication**
    - 15—Journalism and Communication


   


```python
# creating a new column with the new categories based on the values of the 'Course' column

# I first tried the np.where() but it only accepts 3 arguments, so when I wanted to add 3+ arguments, it returned an error

# Define a function for more complex mappings
def map_course(Course):
    if Course in [1, 7]:  
        return '1'
    elif Course in [2, 5]:  
        return '2'
    elif Course in [4, 6, 8]:  
        return '3'
    elif Course in [3, 10, 16]: 
        return '4'
    elif Course in [9, 11, 14, 17]: 
        return '5'
    elif Course in [12, 13]: 
        return '6'
    elif Course == 15:  
        return '7'
    else:
        return 'Other'  # default value if none of the above matches

# Apply the function to create the new column
df['broad_course_type'] = df['Course'].apply(map_course)

# checking
df['broad_course_type'].value_counts()
```

```python
# now you can mess with the range values to see whether increasing the number of trees makes a difference in the MAE of the 
# model's predictions 
results = []
for i in range(1,5):
    # Creating the classifier object.
    RFR = RandomForestClassifier(n_estimators=i,
                                 max_depth=10, 
                                 random_state=42)

    # Fitting the train dataset.
    RFR.fit(X_train,Y_train)

    # Creating the predictions.
    Y_pred = RFR.predict(X_test)

    # Calculate error
    ys_mae = mean_absolute_error(Y_test,Y_pred)
    
    # Appending to the list.
    results.append(ys_mae)
    
print(min(results))

plt.plot(results)

# the line of MAE plateaus after 2 so it suggests there is minimal improvement with more trees. I think its because the data is
# maybe too simple there isnt much complexity. 
```
