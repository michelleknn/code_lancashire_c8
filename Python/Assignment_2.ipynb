<h1>Secondary analysis</h1>

My secondary question: could macro-economic variables predict courses enrolled? 

The current course category is very specific and I'm not interested in whether our variables can predict specific courses but rather the broader category so let's regroup them into 6 broad categories and collapse the evening courses with the normal ones: 

1. **Engineering and Technology**
    - 1—Biofuel Production Technologies
    - 7—Informatics Engineering
  
2. **Arts, Design, and Multimedia**
    - 2—Animation and Multimedia Design
    - 5—Communication Design

3. **Agriculture and Veterinary Sciences**
    - 4—Agronomy
    - 6—Veterinary Nursing
    - 8—Equiniculture
    
4. **Social Sciences and Services**
    - 3—Social Service (evening attendance)
    - 10—Social Service
    - 16—Basic Education

5. **Business and Management**
    - 9—Management
    - 11—Tourism
    - 14—Advertising and Marketing Management
    - 17—Management (evening attendance)
      
6. **Health Sciences**
    - 12—Nursing
    - 13—Oral Hygiene
      
7. **Media and Communication**
    - 15—Journalism and Communication


   


```python
# creating a new column with the new categories based on the values of the 'Course' column

# I first tried the np.where() but it only accepts 3 arguments, so when I wanted to add 3+ arguments, it returned an error

# Define a function for more complex mappings
def map_course(Course):
    if Course in [1, 7]:  
        return '1'
    elif Course in [2, 5]:  
        return '2'
    elif Course in [4, 6, 8]:  
        return '3'
    elif Course in [3, 10, 16]: 
        return '4'
    elif Course in [9, 11, 14, 17]: 
        return '5'
    elif Course in [12, 13]: 
        return '6'
    elif Course == 15:  
        return '7'
    else:
        return 'Other'  # default value if none of the above matches

# Apply the function to create the new column
df['broad_course_type'] = df['Course'].apply(map_course)

# checking
df['broad_course_type'].value_counts()
```

<H3>Visualisation</H3>

```python
# Creating a pie chart of courses from the broad_course_type column 
#Step 1: Calculate the distribution of broad_course_type
broad_course_distribution = df['broad_course_type'].value_counts()

# Step 2: Create the pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(broad_course_distribution, labels=broad_course_distribution.index, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Distribution of Broad Course Type')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Display the pie chart
plt.show()
```

```python

# Step 1: Create a mapping from broad_course_type to meaningful names
broad_course_labels = {
    '1': 'Engineering and Technology\n• Biofuel Production Technologies\n• Informatics Engineering',
    '2': 'Arts, Design, and Multimedia\n• Animation and Multimedia Design\n• Communication Design',
    '3': 'Agriculture and Veterinary Sciences\n• Agronomy\n• Veterinary Nursing\n• Equiniculture',
    '4': 'Social Sciences and Services\n• Basic Education\n• Social Service\n• Social Service (evening attendance)',
    '5': 'Business and Management\n• Management\n• Management (evening attendance)\n• Advertising and Marketing Management\n• Tourism',
    '6': 'Health Sciences\n• Nursing\n• Oral Hygiene',
    '7': 'Media and Communication\n• Journalism and Communication',
    'Other': 'Other'
}

# Step 2: Apply meaningful names to the broad_course_type column
df['broad_course_type_named'] = df['broad_course_type'].map(broad_course_labels)

# Step 3: Calculate the distribution of the broad_course_type
broad_course_distribution = df['broad_course_type_named'].value_counts()

# Step 4: Create the pie chart
# Step 4: Create the pie chart with leader lines
plt.figure(figsize=(10, 10))  # Set the figure size

# Define colors for better visualization (optional)
colors = plt.get_cmap('Set3').colors  # You can customize colors if needed

# Plot the pie chart with leader lines
plt.pie(
    broad_course_distribution,
    labels=broad_course_distribution.index,  # Add the meaningful labels
    autopct='%1.1f%%',  # Add percentages
    startangle=90,  # Start the first slice at the top
    pctdistance=0.85,  # Position the percentage inside the pie slice
    labeldistance=1.1,  # Position the labels outside the pie chart
    wedgeprops={"linewidth": 2, "edgecolor": "w"},  # Draw the borders between slices
    colors=colors  # Optional: Set colors
)

# Add a title
plt.title('Distribution of Broad Course Types', pad=40, fontweight='bold') # pad to create some space between the title and labels as it was overlapping before

# Equal aspect ratio ensures that the pie is drawn as a circle
plt.axis('equal')

# Show the pie chart
plt.show()
```
